```python
# apply
# who is under the age of 20 would be a child
# between 20 and 40 would be youth
# 40 - 60 would be middle aged
# > 60 would be considered old
import numpy as np

np.ceil(25.22) 
```

```python
def categorize_age(age):
    age_whole=int(np.ceil(age))
    if(age<20):
        return "child"
    elif(20 <= age_whole & age_whole < 40):
        return "youth"
    elif(40 <= age_whole & age_whole < 60):
        return "middle aged"
    else:
        return "old"
```

```python
titanic['age_category']= titanic['Age'].apply(categorize_age)
```
