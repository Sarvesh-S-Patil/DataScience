```python
file_handler = open(location,mode)

lines = list(file_handler) 

bikes = pd.DataFrame(lines)
```

```python
item_list = []
for line in lines[1:]:
    split = line.split(',')
    item = split[0]
    if not item in item_list:
        item_list.append(item)
print(item_list)
```

```python
prices = []
for item in item_list:
    total=0
    for line in lines[1:]:
        parts = line.split(',')
        if (parts[0] == item):
            try:
                exc_price = float(parts[7])
                total += exc_price
            except:
                continue 
                
    prices.append(total)
```

```python
print(prices)
```
