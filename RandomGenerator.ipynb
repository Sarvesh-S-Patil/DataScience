```python
rows = int(input("Enter no of rows to be generated : ")) 
```

```python
import numpy as np
```

```python
Student_info = [[0 for i in range(7)] for j in range(rows)]
```

```python
def Random_Matrix_Generator(rows):
    j=1
    i=0
    while(i<rows):
        student_id=j
        first_name= ''.join(random.choices(string.ascii_lowercase,k=5))
        last_name= ''.join(random.choices(string.ascii_lowercase,k=5))
        regular_attendance=random.choices(['Yes','No'])
        standard= random.choices([1,2,3,4,5,6,7,8,9])
        gender= random.choices(['M','F'])
        marks_scored=random.randint(0,100)
        Student_info[i]= student_id,first_name,last_name,regular_attendance,standard,gender,marks_scored
        
        j+=1
        i+=1
        
        
    return Student_info
```

```python
Student_info = Random_Matrix_Generator(rows)
```

```python
Student_info_df = pd.DataFrame(Student_info)
```
